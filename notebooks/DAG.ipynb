# Incremental files ingestion from Azure Blob Storage to Microsoft Fabric Lakehouse using PySpark Notebooks

> <br> This project uses the database provided by the government of Brazil called **Cadastro Geral de Empregados e Desempregados (CAGED)** (General Register of Employed and Unemployed).<br><br>

**Notebook**: DAG  
**Description**: This PySpark notebook just orchestrate all the notebboks in logical sequence due their dependencies. 

```python
# run multiple notebooks with parameters
DAG = {
    "activities": [
        {
            "name": "BlobToLanding", 
            "path": "BlobToLanding",
        },
        {
            "name": "LandingToBronze",
            "path": "LandingToBronze",
            "dependencies": ["BlobToLanding"],
        },
        {
            "name": "BronzeToSilver",
            "path": "BronzeToSilver",
            "dependencies": ["LandingToBronze"],
        },
        {
            "name": "SilverToGold",
            "path": "SilverToGold",
            "dependencies": ["BronzeToSilver"],
        },
    ],
}
notebookutils.notebook.runMultiple(DAG, {"displayDAGViaGraphviz": False})
```
